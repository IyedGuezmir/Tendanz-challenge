```python
from docling.document_converter import DocumentConverter
import os
from pathlib import Path

def parse_pdf(input_doc_path: Path, output_dir: Path):

    md_filepath = output_dir / f"{input_doc_path.stem}-parsed-text.md"
    
    if md_filepath.exists():
        return md_filepath
    
    doc_converter = DocumentConverter()
    conv_res = doc_converter.convert(input_doc_path)

    with md_filepath.open("w", encoding="utf-8") as md_file:
        md_file.write(conv_res.document.export_to_markdown())
    
    return md_filepath
```

```python
def get_chunks_using_markers(src_text: str) -> list[str]:
    """
    Split the source text into chunks using major section markers.
    """
    # Major sections from your document (extended list)
    major_sections = [
        "# QUELQUES DFINITIONS",
        "# VOUS ET VOTRE CONTRAT",
        "# SYNTHSE DES GARANTIES PROPOSES",
        "# PRSENTATION DES GARANTIES",
        "# CE QUE VOTRE CONTRAT NE GARANTIT JAMAIS",
        "# EN CAS DE SINISTRE",
        "# LA VIE DE VOTRE CONTRAT",
        "# LES DCLARATIONS QUE VOUS DEVEZ FAIRE",
        "# LE KILOMTRAGE",
        "# LA COTISATION",
        "# VOTRE INFORMATION",
        "# FICHE D'INFORMATION"
    ]
    
    chunks = []
    current_pos = 0
    
    # Find each major section and split there
    for section in major_sections:
        idx = src_text.find(section, current_pos)
        if idx != -1:
            # Add chunk from current position to start of this section
            if idx > current_pos:
                chunk = src_text[current_pos:idx].strip()
                if chunk:
                    chunks.append(chunk)
            current_pos = idx
    
    # Add the final chunk
    if current_pos < len(src_text):
        final_chunk = src_text[current_pos:].strip()
        if final_chunk:
            chunks.append(final_chunk)
    
    # If no major sections were found (fallback to original logic)
    if not chunks:
        marker = "\n#"
        parts = src_text.split(marker)
        
        if parts[0].strip():
            chunks.append(parts[0].strip())
        
        for part in parts[1:]:
            if part.strip():
                chunks.append(marker + part.strip())
    
    return chunks
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

def split_chunk_with_langchain(chunk: str, chunk_size: int = 5000, chunk_overlap: int = 200) -> list[str]:
    """
    Split chunk while keeping tables and their explanations together.
    Only split at subheadings or if chunk is very large.
    """
    # Step 1: Split only at minor headings
    splitter = RecursiveCharacterTextSplitter(
        separators=["\n##"],  # only split by minor headings
        chunk_size=chunk_size,
        chunk_overlap=chunk_overlap
    )
    chunks = splitter.split_text(chunk)
    
    # Step 2: Merge small orphan chunks with previous chunk
    final_chunks = []
    for c in chunks:
        if final_chunks and len(c) < 300:  # threshold for "small orphan"
            final_chunks[-1] += "\n\n" + c
        else:
            final_chunks.append(c)
    
    return final_chunks
```

```python
major_chunks = get_chunks_using_markers(md_txt)
# Step 2: Split large chunks further using LangChain
all_chunks = []
for chunk in major_chunks:
    all_chunks.extend(split_chunk_with_langchain(chunk, chunk_size=5000, chunk_overlap=200))

print(f"Total chunks: {len(all_chunks)}")  # should be ~4050 chunks for your 68-page PDF
```
